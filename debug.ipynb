```python
# importar dados de uma unica ação da Procter & Gamble (PG) usando yfinance 
import yfinance as yf

# Baixar os dados da PG usando yfinance
PG = yf.download('PG', start='1995-1-1')
PG
```

```python
# importar dados de multipals ações usando yfinance 
import yfinance as yf
import pandas

tickers = ['PG', 'MSFT', 'T', 'F', 'GE']
new_data = pandas.DataFrame(columns=tickers)
for i in tickers:
    new_data[i] = yf.download(i, start='1995-1-1')['Adj Close']
new_data.tail()
```
